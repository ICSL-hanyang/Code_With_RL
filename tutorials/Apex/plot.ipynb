```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
Apex_SAC = np.loadtxt("return.txt", delimiter=',')
```

```python
plt.figure(figsize=(10,5))
plt.title("Average Reward")
plt.grid(True)

plt.xlabel(" Time (minute) ")
plt.ylabel("Average reward (10 time)")

plt.plot(Apex_SAC, "r", linewidth=2.0, label="Apex SAC Average Reward")
plt.legend(loc="best", shadow=True)

plt.savefig('result.png')

plt.show()
```
