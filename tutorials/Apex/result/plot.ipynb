```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
apex_sac = np.loadtxt("apex_sac.txt", delimiter=',')
SAC_10_step = np.loadtxt("sac.txt", delimiter=',')
```

```python
plt.figure(figsize=(10,5))
plt.title("Average Reward")
plt.grid(True)

plt.xlabel(" Time ")
plt.ylabel("Average reward (10 time)")

plt.plot(apex_sac, "r", linewidth=2.0, label="Apex SAC Average Reward")
plt.plot(SAC_10_step, "b", linewidth=2.0, label="SAC Average Reward")

plt.legend(loc="best", shadow=True)

plt.savefig('result.png')

plt.show()
```
