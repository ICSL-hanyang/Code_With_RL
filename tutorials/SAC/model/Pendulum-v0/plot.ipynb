```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
SAC_10_step = np.loadtxt("step_return.txt", delimiter=',')
SAC_episode = np.loadtxt("episode_return.txt", delimiter=',')
```

```python
plt.figure(figsize=(10,5))
plt.title("Average Reward")
plt.grid(True)

plt.xlabel(" Time ")
plt.ylabel("Average reward (10 time)")
plt.plot(SAC_episode, "r", linewidth=2.0, label="Apex SAC Average Reward")
plt.plot(SAC_10_step, "b", linewidth=2.0, label="Apex SAC Average Reward")
plt.legend(loc="best", shadow=True)

# plt.savefig('result.png')

plt.show()
```
